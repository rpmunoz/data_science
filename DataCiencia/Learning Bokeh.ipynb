```python
from bokeh.plotting import figure, show
from bokeh.charts import Donut, show, output_file
from bokeh.charts.utils import df_from_json
from bokeh.sampledata.olympics2014 import data
from bokeh.io import output_notebook
from bokeh.models import HoverTool, BoxSelectTool
from collections import OrderedDict

import pandas as pd
```

```python
# utilize utility to make it easy to get json/dict data converted to a dataframe
df = df_from_json(data)
df.head()
```

```python
# filter by countries with at least one medal and sort by total medals
df = df[df['total'] > 8]
df = df.sort_values(by="total", ascending=False)
df = pd.melt(df, id_vars=['abbr'],
             value_vars=['bronze', 'silver', 'gold'],
             value_name='medal_count', var_name='medal')

df.head()
```

```python
output_notebook()

#hover = HoverTool( tooltips=[("abbr", '@x'),("(x,y)", "($x, $y)"),("desc", "@medal_count"),] )

#TOOLS = [BoxSelectTool(), HoverTool()]
# original example
p = Donut(df, title="Medallas olímpicas", label=['abbr', 'medal'], values='medal_count',
          text_font_size='8pt', height=600, width=600, tools=[HoverTool()])

hover = p.select(dict(type=HoverTool))
hover.tooltips = """<div>Value of ID: @medal</div>
<div>Value of Total: @medal_count</div>
"""

##hover.tooltips = [('Value of ID',' $abbr'),('Value of Desc',' @Desc')]
#p.add_tools(HoverTool(tooltips=[
#    ("Organisation Name", "@OrganisationName")
#]))

show(p)
```

```python
import bokeh.plotting as bk
import pandas as pd
import numpy as np
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.palettes import brewer
```

```python
rad2deg = (180 / np.pi)
deg2rad = 1 / rad2deg

y_mins = [0.5] * 3
y_maxes = [1] * 3
starts = pd.Series([30, 45, 90])*deg2rad
stops = pd.Series([40, 80, 120])*deg2rad
colors = brewer['Spectral'][len(start_angles)]
name=['PolaPig','PepaPig','Oink']

df = pd.DataFrame(
                    {
                        'x': 0,
                        'y': 0,
                        'ymin': y_mins, 
                        'ymax': y_maxes,
                        'starts': starts,
                        'stops': stops,
                        'colors': colors,
                        'name': name 
                    }
                )

output_notebook()

df_s = ColumnDataSource(df)
TOOLS = "hover"
p = bk.figure(title='Proyectos Fondecyt regular',x_range=(-1.1,1.1), y_range=(-1.1,1.1), width=600, height=600, tools=TOOLS)

p.title.align='center'
p.title.text_font='Roboto'
p.title.text_font_size='14pt'
p.title.text_alpha=0.7
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

p.annular_wedge(x='x', y='y', 
                inner_radius='ymin',
                outer_radius='ymax',
                start_angle='starts',
                end_angle='stops',
                color='colors',
                direction="anticlock",
                source=df_s
                )

hover = p.select(dict(type=HoverTool))
hover.tooltips = [("Nombre", "@name")]
#hover.snap_to_data = 'anchor'
bk.show(p)
```
